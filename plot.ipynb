```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('log.csv')
```

```python
data
```

## По столбцам

```python
plt.title('Время работы')
plt.ylabel('Время')
plt.xlabel('Размер матриц')
plt.plot(data.loc[:9,'size'], data.loc[:9,' time_col'])
plt.show()
```

```python
plt.title('Ускорение')
plt.ylabel('Ускорение')
plt.xlabel('Размер матриц')
plt.plot(data.loc[:9,'size'], data.loc[10:,' time_col'].values / data.loc[:9,' time_col'].values)
plt.show()
```

```python
plt.title('Эффективность')
plt.ylabel('Эффективность')
plt.xlabel('Размер матриц')
plt.plot(data.loc[:9,'size'], (data.loc[10:,' time_col'].values / data.loc[:9,' time_col'].values)/4)
plt.show()
```
